```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Kaggle.csv')
```

```python
df
```

```python
df.plot()
```

```python
df.drop(['Solids'], axis = 1, inplace = True)
df
```

```python
ax = df.plot(figsize=(15,8), title = "Kaggle Water Report")
ax.set_xlabel('Sample size')
ax.set_xlabel('Parameter Reading')
```

```python
df.loc[1:200].dropna().plot(figsize=(15,8), title = "200 Samples", kind = 'line').set_xlabel("Number of samples")
```

```python
df.loc[1:200].dropna().plot(figsize=(15,8), title = "200 Samples", kind = 'hist',stacked = True).set_xlabel("Number of samples")
```
